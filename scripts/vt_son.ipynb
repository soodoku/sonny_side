```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
import time
```

```python
options = webdriver.ChromeOptions()
# Using SOCKS Proxy (Optional)
options.add_argument("--proxy-server=socks5://127.0.0.1:8081")
browser = webdriver.Chrome('f:/utils/chromedriver/chromedriver.exe', options=options)
browser.get("https://www.vtsosonline.com/online/BusinessInquire")
```

```python
elem = browser.find_element_by_id("rbContains")
elem.click()
entity = browser.find_element_by_id("txtBusinessName")
entity.clear()
entity.send_keys("son")
search = browser.find_element_by_id("btnSearch")
search.click()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

df = None
page = 1
```

```python
while True:
    try:
        print('page = %d' % page)
        elem = browser.find_element_by_id("txtCommonPageNo")
        elem.clear()
        elem.send_keys(str(page))
        elem = browser.find_element_by_id("lkGoPage")
        elem.click()
    except Exception as e:
        print(e)
        break
    time.sleep(1)
    try:
        element = WebDriverWait(browser, 180).until(
            EC.invisibility_of_element_located((By.ID, "process"))
        )
    except Exception as e:
        print(e)
        break
    time.sleep(1)        
    dfs = pd.read_html(browser.page_source, attrs={'id': 'xhtml_grid'})
    if df is None:
        df = dfs[0]
    else:
        df = pd.concat([df, dfs[0]])
    page += 1        
```
